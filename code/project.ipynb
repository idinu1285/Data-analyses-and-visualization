### Let's start doing a short data analysis of data frame

### Displaying the titles of movies which starts with 'A' and  ends with 'n'

```python
titlu_film = filme[(filme['Titlu'].str.startswith('A') & filme['Titlu'].str.endswith('n'))]['Titlu']
titlu_film
```

### Sorting in descending order of the movie titles according to the year of production (the first 15 movie titles, respectively the last 15 movie titles)

```python
sorting_movies_p15 = filme[['Titlu', 'An']].sort_values('An', ascending=False).head(15)
sorting_movies_u15 = filme[['Titlu', 'An']].sort_values('An', ascending=False).tail(15)
```

```python
sorting_movies_p15
```

```python
sorting_movies_u15
```

### Displaying movie titles that were produced in the United States and were produced in 20th Century Fox production studios

```python
country_movie = filme[(filme['Tara'] == 'United States of America') & (filme['Studio'] == '20th Century Fox')][['Titlu','Studio','Tara']]
country_movie
```

### Displaying movies with the lowest estimated budget, respectively the highest estimated budget

```python
def low_high_budget(column):
    
    index_low = filme[column].idxmin()
    index_high = filme[column].idxmax()
    high = pd.DataFrame(filme.loc[index_high, :])
    low= pd.DataFrame(filme.loc[index_low, :])
    return pd.concat([high, low], axis=1)  

low_high_budget('Buget')
```

### Displaying the titles, years of production and actors of movies like Thriller or Action genre

```python
genre_movie = filme[(filme['Gen'] == 'Thriller') | (filme['Gen'] == 'Action')][['Titlu', 'Gen']]
genre_movie
```

### Displaying the names of the directors who directed films with a duration between 180 and 240 minutes

```python
movie_duration = filme[(filme['Durata'] >= 180) & (filme['Durata'] <= 240)][[ 'Regizor','Durata']]
movie_duration
```

### Displaying  the average duration for each movie type

```python
duration_mean = filme[['Durata', 'Gen']].groupby('Gen').mean()
duration_mean
```

### Displaying the titles of the movies with the highest rating, respectively the lowest for each genre

```python
rating_high_genre = filme.sort_values('Rating', ascending = False).groupby('Gen')['Titlu','Rating'].max()
rating_low_genre =  filme.sort_values('Rating', ascending = False).groupby('Gen')['Titlu','Rating'].min()
```

```python
rating_high_genre
```

```python
rating_low_genre
```

### Displaying the names of the actors who are the protagonists of the movies that have the main language Romanian or Spanish 

```python
language_actor = filme[(filme['Limba'] == 'Romanian') | (filme['Limba'] == 'Spanish')]['Actor_principal']
language_actor
```

### Finally let's make some data visualizations
### Graphical representation of the distributions of durations, ratings and production years

```python
filme['Durata'].plot.hist(title='Distribution of movie durations', color = 'b')
plt.xlabel('Movie duration')
plt.ylabel('Movies total')
plt.show()
filme['Rating'].plot.hist(title='Distribution of movie ratings', color = 'b')
plt.xlabel('Movie rating')
plt.ylabel('Movies total')
plt.show()
filme['An'].plot.hist(title='Distribution of production years', color = 'b')
plt.xlabel('Production year')
plt.ylabel('Movies total')
plt.show()
```

### Graphical representation of the correlation between the year of production and the duration of the movies

```python
ax = sns.regplot(x=filme['An'], y=filme['Durata'],color='r')
ax.set_title('Correlation between production year and duration',fontsize=16)
ax.set_xlabel('Production year',fontsize=13)
ax.set_ylabel('Movie udaration[min]',fontsize=13)
sns.set(rc={'figure.figsize':(11,8)})
sns.set_style('ticks', {'xtick.major.size': 8, 'ytick.major.size': 8})
plt.show()
```

### Graphical representation of the correlation between the production year and  budget of movies

```python
ax = sns.regplot(x=filme['An'], y=filme['Buget'],color='r')
ax.set_title('Correlation between production year and estimated budget',fontsize=16)
ax.set_xlabel('Production year',fontsize=13)
ax.set_ylabel('Estimated budget [USD]',fontsize=13)
sns.set(rc={'figure.figsize':(11,8)})
sns.set_style('ticks', {'xtick.major.size': 8, 'ytick.major.size': 8})
plt.show()
```

### Graphical representation of the correlation between the production year and rating of movies

```python
ax = sns.regplot(x=filme['An'], y=filme['Rating'],color='r')
ax.set_title('Correlation between production year and rating',fontsize=16)
ax.set_xlabel('Production year',fontsize=13)
ax.set_ylabel('Movie rating',fontsize=13)
sns.set(rc={'figure.figsize':(11,8)})
sns.set_style('ticks', {'xtick.major.size': 8, 'ytick.major.size': 8})
plt.show()
```

### Displaying a boxplot graph which represents the correlation between the type of movies and its duration

```python
sns.boxplot(data=filme, x = 'Tip', y = 'Durata')
plt.title('Correlation between type and duration', fontsize = 16)
sns.set_style('ticks', {'xtick0.major.size': 8, 'ytick.major.size': 8})
plt.show()
```

### Which is the top 20 best rated movies?

```python
top_20_rating = filme.nlargest(20,'Rating')
sns.barplot(x='Rating', y='Titlu', data=top_20_rating, palette='Blues')
plt.xlabel('Movie rating')
plt.ylabel('Movie title')
sns.set(rc={'figure.figsize':(11,8)})
sns.set_style('ticks', {'xtick.major.size': 8, 'ytick.major.size': 8})
plt.title('Top 20 best rated movies', fontsize = 18)
plt.show()
```

### Which is the top 20 longest movies?

```python
top_20_durata = filme.nlargest(20,'Durata')
sns.barplot(x='Durata', y='Titlu', data=top_20_durata, palette='Greens')
plt.xlabel('Movie duration [min]')
plt.ylabel('Movie title')
sns.set(rc={'figure.figsize':(11,7)})
sns.set_style('ticks', {'xtick.major.size': 8, 'ytick.major.size': 8})
plt.title('Top 20 longest movies', fontsize = 18)
plt.show()
```

### Displaying a lmplot graph which shows the relationship between rating, duration and main language of movie

```python
sns.set_style('white')
sns.lmplot(data=filme, x='Rating', y='Durata', fit_reg=False, hue='Limba', size=6, aspect=1)
plt.title('Relation rating - duration - language', fontsize = 14)
plt.xlabel('Movie rating')
plt.ylabel('Movie duration [min]')
plt.show()
```

### Displaying a lmplot grapg which shows the relationship between rating, duration and country  of movie

```python
sns.set_style('white')
sns.lmplot(data=filme, x='Rating', y='Durata', fit_reg=False, hue='Tara', size=6, aspect=1)
plt.title('Relation rating - duration - country', fontsize = 14)
plt.xlabel('Movie rating')
plt.ylabel('Movie duration [min]')
sns.color_palette('dark')
plt.show()
```
